```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Csv/741999.'
water_df = pd.read_csv(path)
water_df.head()
```

```python
y = water_df['Bicarb'].values.reshape(-1, 1)
x = water_df['pH'].values.reshape(-1, 1)
```

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python

y_pred = reg.predict(x_test)
```

```python


from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python

print("Enter pH value: ")
pH = float(input())
output = reg.predict([[pH]])
print("The Bicarbonate value", output)
```
